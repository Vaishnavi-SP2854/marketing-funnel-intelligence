```python
import pandas as pd
import matplotlib.pyplot as plt 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/marketing_funnel_clean.csv")
```

```python
conversion_rate = df["Converted"].mean()
print("Overall Conversion Rate:", round(conversion_rate,3))
```

```python
segment_conv = df.groupby("business_segment")["Converted"].mean().sort_values(ascending=False)
segment_conv
```

```python
segment_conv.plot(kind="bar", figsize=(8,4))
plt.title("Conversion Rate by Business Segment")
plt.show()
```
