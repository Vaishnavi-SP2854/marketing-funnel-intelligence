```python
import pandas as pd

df = pd.read_csv("../data/raw/marketing_funnel_raw.csv")
```

```python
df["won_date"] = pd.to_datetime(df["won_date"], errors="coerce")
```

```python
df["Converted"] = df["won_date"].notnull().astype(int)
```

```python
import numpy as np

def clean_stock(value):
    if pd.isna(value):
        return np.nan
    if isinstance(value, str):
        if "-" in value:
            low, high = value.split("-")
            return (float(low) + float(high)) / 2
        elif "+" in value:
            return float(value.replace("+",""))
        elif value.lower() == "unknown":
            return np.nan
    return value

df["average_stock"] = df["average_stock"].apply(clean_stock)
```

```python
df["average_stock"] = pd.to_numeric(df["average_stock"], errors="coerce")
```

```python
df["average_stock"].fillna(df["average_stock"].median(), inplace=True)
```

```python
df["declared_monthly_revenue"].fillna(0, inplace=True)
df["declared_product_catalog_size"].fillna(0, inplace=True)
df["average_stock"].fillna(df["average_stock"].median(), inplace=True)
```

```python
df["has_company"] = df["has_company"].map({True:1, False:0})
df["has_gtin"] = df["has_gtin"].map({True:1, False:0})
```
